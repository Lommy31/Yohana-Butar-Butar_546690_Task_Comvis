# 1st Assigment - Computer Vision

#Brightened Image

```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
from IPython.display import clear_output

uploaded = files.upload()
clear_output()
filename = list(uploaded.keys())[0]

image = cv2.imread(filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

bright_img = cv2.convertScaleAbs(image, alpha=1, beta=50)

def compare(img1, img2, title1="Image 1", title2="Image 2"):
  plt.figure(figsize=(12,5))
  plt.subplot(1,2,1)
  plt.title(title1)
  plt.imshow(img1)
  plt.axis("off")

  plt.subplot(1,2,2)
  plt.title(title2)
  plt.imshow(img2)
  plt.axis("off")

  plt.show()

compare(image, bright_img, "Original Image", "Brightened Image")
```

#Darkened Image

```python
uploaded = files.upload()
clear_output()
filename = list(uploaded.keys())[0]

image = cv2.imread(filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

dark_img = cv2.convertScaleAbs(image, alpha=1, beta=-50)

compare(image, dark_img, "Original Image", "Darkened Image")
```

#Noisy Image

```python
def compare(img1, img2, title1="Image 1", title2="Image 2"):
    plt.figure(figsize=(12,5))

    plt.subplot(1,2,1)
    plt.title(title1)
    plt.imshow(img1)
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.title(title2)
    plt.imshow(img2)
    plt.axis("off")

    plt.show()

uploaded = files.upload()
clear_output()
filename = list(uploaded.keys())[0]

image = cv2.imread(filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

denoise_img = cv2.medianBlur(image,7)

compare(image, denoise_img, "Original Image", "Median Filter Image")
```

#Contrast Image

```python
uploaded = files.upload()
clear_output()
filename = list(uploaded.keys())[0]

image = cv2.imread(filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

brightContrast_img = cv2.convertScaleAbs(image, alpha=1.4, beta=0)
lowContrast_img = cv2.convertScaleAbs(image, alpha=1, beta=0)

def compare(img1, img2, img3, title1="Original", title2="Bright", title3="Low Contrast"):
    plt.figure(figsize=(15,5))

    for i,(img,title) in enumerate([(img1,title1),(img2,title2),(img3,title3)]):
        plt.subplot(1,3,i+1)
        plt.imshow(img)
        plt.title(title)
        plt.axis("off")

    plt.show()

compare(image, brightContrast_img, lowContrast_img, "Original Image", "Bright Contrast Image", "Low Contrast Image")
```

#Low Resolution Image

```python
uploaded = files.upload()
clear_output()
filename = list(uploaded.keys())[0]

image = cv2.imread(filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

h, w = image.shape[:2]

mask = np.zeros((h,w), np.uint8)
cv2.line(mask, (0,220), (w,260), 255, 50)

restored = cv2.inpaint(image, mask, 7, cv2.INPAINT_TELEA)

restored = cv2.medianBlur(restored,5)

kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])

restored = cv2.filter2D(restored,-1,kernel)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Restored Image")
plt.imshow(image)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Original Image")
plt.imshow(restored)
plt.axis("off")

plt.show()
```
